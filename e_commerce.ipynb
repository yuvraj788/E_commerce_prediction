```python
x = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values
```

```python
print(x)
```

### *Split Data*

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(x_train)
```

### *Scaler*

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
#from sklearn.model_selection import train_test_split # Importing train_test_split from sklearn
from sklearn.tree import DecisionTreeRegressor # Importing DecisionTreeRegressor from sklearn
from sklearn.ensemble import RandomForestRegressor # Importing RandomForestRegressor from sklearn
from sklearn.linear_model import LinearRegression # Importing LinearRegression from sklearn
from sklearn.metrics import mean_squared_error # Importing mean_squared_error from sklearn
import xgboost as xgb # Importing xgboost from xgboost
import lightgbm as lgb # Importing lightgbm from lightgbm
from sklearn.model_selection import GridSearchCV # Importing GridSearchCV from sklearn
```

### *Linear_Regression*

```python
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
lr_preds_test = lr_model.predict(x_test)
lr_train_score = lr_model.score(x_train, y_train)
lr_test_score = lr_model.score(x_test, y_test)
lr_rmse = mean_squared_error(y_test, lr_preds_test)**(1/2)

print(f"Accuracy score for LinearRegression on the train set: {lr_train_score}")
print(f"Accuracy score for LinearRegression on the test set: {lr_test_score}")
print(f"RMSE score for LinearRegression on the test set: {lr_rmse}")
```

### *DecisionTree_Regressor*

```python
dt_model = DecisionTreeRegressor(max_depth=5, random_state=42, min_samples_leaf=5, min_samples_split=5)
dt_model.fit(x_train, y_train)
dt_preds_test = dt_model.predict(x_test)
dt_train_score = dt_model.score(x_train, y_train)
dt_test_score = dt_model.score(x_test, y_test)
dt_rmse = mean_squared_error(y_test, dt_preds_test)**(1/2)

print(f"Accuracy score for DecisionTreeRegressor on the train set: {dt_train_score}")
print(f"Accuracy score for DecisionTreeRegressor on the test set: {dt_test_score}")
print(f"RMSE score for DecisionTreeRegressor on the test set: {dt_rmse}")
```

### *RandomForest_Regressor*

```python
rf_model = RandomForestRegressor(n_estimators=100, max_depth=5, random_state=42, min_samples_leaf=5, min_samples_split=5)
rf_model.fit(x_train, y_train)
rf_preds_test = rf_model.predict(x_test)
rf_train_score = rf_model.score(x_train, y_train)
rf_test_score = rf_model.score(x_test, y_test)
rf_rmse = mean_squared_error(y_test, rf_preds_test)**(1/2)

print(f"Accuracy score for RandomForestRegressor on the train set: {rf_train_score}")
print(f"Accuracy score for RandomForestRegressor on the test set: {rf_test_score}")
print(f"RMSE score for RandomForestRegressor on the test set: {rf_rmse}")
```

### *GridSearchCV with RandomForestRegressor*

```python
rf_grid_model = GridSearchCV(RandomForestRegressor(random_state=42), param_grid={'n_estimators': [100, 200, 300, 400, 500], 'max_depth': [5, 10, 15, 20, 25]}, cv=5)
rf_grid_model.fit(x_train, y_train)
rf_grid_preds_test = rf_grid_model.predict(x_test)
rf_grid_train_score = rf_grid_model.score(x_train, y_train)
rf_grid_test_score = rf_grid_model.score(x_test, y_test)
rf_grid_rmse = mean_squared_error(y_test, rf_grid_preds_test)**(1/2)

print(f"Accuracy score for GridSearchCV with RandomForestRegressor on the train set: {rf_grid_train_score}")
print(f"Accuracy score for GridSearchCV with RandomForestRegressor on the test set: {rf_grid_test_score}")
print(f"RMSE score for GridSearchCV with RandomForestRegressor on the test set: {rf_grid_rmse}")
```

### *XGBRegressor*

```python
xgb_model = xgb.XGBRegressor(n_estimators=100, max_depth=5, random_state=42)
xgb_model.fit(x_train, y_train)
xgb_preds_test = xgb_model.predict(x_test)
xgb_train_score = xgb_model.score(x_train, y_train)
xgb_test_score = xgb_model.score(x_test, y_test)
xgb_rmse = mean_squared_error(y_test, xgb_preds_test)**(1/2)

print(f"Accuracy score for XGBRegressor on the train set: {xgb_train_score}")
print(f"Accuracy score for XGBRegressor on the test set: {xgb_test_score}")
print(f"RMSE score for XGBRegressor on the test set: {xgb_rmse}")
```

### *GridSearchCV with XGBRegressor*

```python
xgb_grid_model = GridSearchCV(xgb.XGBRegressor(random_state=42), param_grid={'n_estimators': [100, 200, 300, 400, 500], 'max_depth': [5, 10, 15, 20, 25]}, cv=5)
xgb_grid_model.fit(x_train, y_train)
xgb_grid_preds_test = xgb_grid_model.predict(x_test)
xgb_grid_train_score = xgb_grid_model.score(x_train, y_train)
xgb_grid_test_score = xgb_grid_model.score(x_test, y_test)
xgb_grid_rmse = mean_squared_error(y_test, xgb_grid_preds_test)**(1/2)

print(f"Accuracy score for GridSearchCV with XGBRegressor on the train set: {xgb_grid_train_score}")
print(f"Accuracy score for GridSearchCV with XGBRegressor on the test set: {xgb_grid_test_score}")
print(f"RMSE score for GridSearchCV with XGBRegressor on the test set: {xgb_grid_rmse}")
```
